```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

```python
data = pd.read_csv('ZomatoChennai.csv')
data
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data.replace(to_replace = ['None','Invalid','Does not offer Delivery','Does not offer Dining','Not enough Delivery Reviews','Not enough Dining Reviews'], value =np.nan,inplace=True)
data.isna().sum()
```

```python
data1 = data.drop(['Location','Top Dishes', 'Delivery Rating', 'Delivery Rating Count','Features'],axis=1,inplace=False)
data1.isna().sum()
```

```python
df = data1.dropna(subset=["Dining Rating","Dining Rating Count"],axis = 0,inplace = False)
df.reset_index(drop=True, inplace=True)
df.isna().sum()
```
