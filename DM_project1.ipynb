```python
lst = data[data['Price for 2'] == ' Kodambakkam'].index.tolist()
data.drop(lst,axis = 0, inplace=True)
```

```python
lst = data[data['Price for 2'] == ' 2nd Avenue'].index.tolist()
data.drop(lst,axis = 0, inplace=True)
```
